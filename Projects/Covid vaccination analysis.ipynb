# Sum  of the vaccine administered

```python
yearwise=data_india.groupby(["Date"]).agg({"Total Covaxin Administered":"sum","Total CoviShield Administered":"sum","Total Sputnik V Administered":"sum","AEFI":"sum"})
yearwise
```

```python
covaxin=data_india["Total Covaxin Administered"].sum().round()
covishield=data_india["Total CoviShield Administered"].sum().round()
sputnik=data_india["Total Sputnik V Administered"].sum().round()
AEFI=data_india["AEFI"].sum().round()
vaccines=[covaxin,covishield,sputnik,AEFI]
vacinscatgry=["Total Covaxin Administered","Total CoviShield Administered","Total Sputnik V Administered","AEFI"]
plt.figure(figsize=(8,5))
plt.plot(vaccines,vacinscatgry,marker='o',color='g')
plt.title("Vaccines Administered",fontsize=15)
plt.show()
```

### Here,we can see CoviShield is given more,then Covaxin is in the second position,Sputnik is in the last position.

# Sum of age wise vaccinated

```python
age18_45=data_india["18-45 years (Age)"].sum().round()
age45_60=data_india["45-60 years (Age)"].sum().round()
age60_abv=data_india["60+ years (Age)"].sum().round()
totalvacintd=data_india["Total Individuals Vaccinated"].sum().round()
ages=[age18_45,age45_60,age60_abv,totalvacintd]
agecatgry=["18-45 years (Age)","45-60 years (Age)","60+ years (Age)","Total Individuals Vaccinated"]
plt.figure(figsize=(8,5))
plt.plot(ages,agecatgry,marker='s',color='r')
plt.title("Age Wise Vaccinated",fontsize=15)
plt.show()
```

### Here,we can see age beween 45 and 60 are vaccinated more.Then age above 60 are in the second position.

### Age between 18 and 45 are vaccinated least.
