## Building Model

```python
X=df.drop(['diagnosis'],axis=1)
y = df['diagnosis']
```

```python
X[:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=40)
```

### Models and finding out the Best one

#### Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()

model1=lr.fit(X_train,y_train)
prediction1=model1.predict(X_test)
```

```python
prediction1
```

```python
d1=pd.DataFrame({"Actual":y_test,"Predicted":prediction1})
d1
```

```python
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,prediction1)
cm
```

```python
sns.heatmap(cm,annot=True)
plt.show()
```

```python
TP=cm[0][0]
TN=cm[1][1]
FN=cm[1][0]
FP=cm[0][1]
print('Testing Accuracy:',(TP+TN)/(TP+TN+FN+FP))
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,prediction1)
```

#### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dtc=DecisionTreeClassifier()
model2=dtc.fit(X_train,y_train)
prediction2=model2.predict(X_test)
```

```python
prediction2
```

```python
d2=pd.DataFrame({"Actual":y_test,"Predicted":prediction1})
d2
```

```python
cm2= confusion_matrix(y_test,prediction2)
cm2
```

```python
sns.heatmap(cm2,annot=True)
plt.show()
```

```python
accuracy_score(y_test,prediction2)
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier

rfc=RandomForestClassifier()
model3 = rfc.fit(X_train, y_train)
prediction3 = model3.predict(X_test)
```

```python
d3=pd.DataFrame({"Actual":y_test,"Predicted":prediction1})
d3
```

```python
cm3=confusion_matrix(y_test, prediction3)
cm3
```

```python
sns.heatmap(cm3,annot=True)
plt.show()
```

```python
accuracy_score(y_test, prediction3)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, prediction1))

print(classification_report(y_test, prediction2))

print(classification_report(y_test, prediction3))
```

#  Conclusion

##### We are getting the best accuracy with RandomForestClassifier which is 97%  , the model is predicting with 97% accuracyon our test data.
#### Random forest shows better performance based on recall scores, which means more malignant tumours were predicted correctly,
